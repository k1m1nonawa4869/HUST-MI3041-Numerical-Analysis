```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
f = lambda x: np.e**x - np.cos(2*x)

# Define the range
x = np.linspace(-1, 1, 1000)

# Compute the function values
y = f(x)

# Plot the graph
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-2, 2) 
plt.axhline(y=0, color='red', linestyle='--')
#plt.axvline(x=0, color='red', linestyle='--')
plt.grid(True)
plt.show()
```
